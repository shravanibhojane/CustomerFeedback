```python
from transformers import pipeline

model_path ="distilbert-base-uncased-finetuned-sst-2-english"

pipe = pipeline(
    model = model_path,
    tokenizer = model_path,
    return_all_scores = True,
    device = "cuda:0",
    truncation=True,
    max_length=512
)
```

```python
from tqdm import tqdm
from transformers.pipelines.pt_utils import KeyDataset
import numpy as np

batch_size = 128
y_pred = []

for i in tqdm(range(0, len(df), batch_size)):
    batch = df["reviews.text"][i : i + batch_size].tolist()
    results = pipe(batch)
    batch_preds = [np.argmax([sent["score"] for sent in res]) for res in results]
    y_pred.extend(batch_preds)
```

```python
y_pred[1000]
```

```python
from sklearn.metrics import classification_report

def evaluate_performance(y_true, y_pred):
    print(classification_report(y_true, y_pred, target_names=["Negative Review", "Positive Review"]))

# Assuming 'reviews.rating' contains numerical ratings, you can create sentiment labels
# This assumes ratings >= 4 are positive, otherwise negative
df["sentiment"] = (df["reviews.rating"] >= 4).astype(int)

# Use the new 'sentiment' column as ground truth
y_true = df["sentiment"].tolist()

# Evaluate model performance
evaluate_performance(y_true, y_pred)
```

```python
import re
from collections import Counter

feedback_texts = df["reviews.text"].dropna().astype(str)

negative_words = [
    "bad", "slow", "broken", "error", "problem", "issues", "waste", "poor",
    "not", "worst", "disappointed", "difficult", "hard", "failed", "crash", "freeze"
]

# Tokenize and filter words related to complaints
all_words = " ".join(feedback_texts).lower()
filtered_words = [word for word in re.findall(r'\b\w+\b', all_words) if word in negative_words]

# Generate a word frequency count
word_freq = Counter(filtered_words)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="Reds").generate_from_frequencies(word_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Frequent Issues in Customer Feedback", fontsize=14)
plt.show()
```
